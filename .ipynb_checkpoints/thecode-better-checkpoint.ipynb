```python
from sklearn.utils import resample


welfake_path = "/home/hatem/.cache/kagglehub/datasets/nitaisatapathy/welfake-dataset/versions/1/WELFake_Dataset.csv"
welfake_raw = pd.read_csv(welfake_path)

welfake_df = welfake_raw.copy()
welfake_df = welfake_df.dropna(subset=["text"])
welfake_df["title"] = welfake_df["title"].fillna("")
welfake_df["content"] = "Title: " + welfake_df["title"] + " [SEP] Body: " + welfake_df["text"]
welfake_df = welfake_df[["content", "label"]].rename(columns={"content": "text"})


print(f"WELFake: {len(welfake_df)} examples")
print(welfake_df["label"].value_counts())
```

```python
liar_raw = load_dataset("chengxuphd/liar2")

liar_label_map = {
    0: 1,     # pants-fire  → Fake
    1: 1,     # false       → Fake
    2: 1,     # barely-true → Fake
    3: None,  # half-true   → DROP (ambiguous)
    4: 0,     # mostly-true → Real
    5: 0,     # true        → Real
}

def process_liar(example):
    new_label = liar_label_map[example["label"]]
    # Combine statement + justification for richer context
    justification = example["justification"] or ""
    statement     = example["statement"]     or ""
    text = "Claim: " + statement + " [SEP] Context: " + justification
    return {"text": text, "label": new_label}

# Process all splits and combine
liar_train = liar_raw["train"].map(process_liar)
liar_val   = liar_raw["validation"].map(process_liar)
liar_test  = liar_raw["test"].map(process_liar)

# Combine all splits into one dataframe
liar_combined = pd.concat([
    pd.DataFrame({"text": liar_train["text"], "label": liar_train["label"]}),
    pd.DataFrame({"text": liar_val["text"],   "label": liar_val["label"]}),
    pd.DataFrame({"text": liar_test["text"],  "label": liar_test["label"]}),
], ignore_index=True)

# Drop half-true (None labels)
liar_df = liar_combined.dropna(subset=["label"])
liar_df["label"] = liar_df["label"].astype(int)

print(f"\nLIAR2 after processing: {len(liar_df)} examples")
print(liar_df["label"].value_counts())
```

```python
fakenews_raw = load_dataset("GonzaloA/fake_news")

fakenews_train = pd.DataFrame({
    "title": fakenews_raw["train"]["title"],
    "text":  fakenews_raw["train"]["text"],
    "label": fakenews_raw["train"]["label"]
})
fakenews_val = pd.DataFrame({
    "title": fakenews_raw["validation"]["title"],
    "text":  fakenews_raw["validation"]["text"],
    "label": fakenews_raw["validation"]["label"]
})
fakenews_test = pd.DataFrame({
    "title": fakenews_raw["test"]["title"],
    "text":  fakenews_raw["test"]["text"],
    "label": fakenews_raw["test"]["label"]
})

fakenews_combined = pd.concat(
    [fakenews_train, fakenews_val, fakenews_test],
    ignore_index=True
)
fakenews_combined = fakenews_combined.dropna(subset=["text"])
fakenews_combined["title"] = fakenews_combined["title"].fillna("")
fakenews_combined["text"]  = (
    "Title: " + fakenews_combined["title"] +
    " [SEP] Body: " + fakenews_combined["text"]
)
fakenews_df = fakenews_combined[["text", "label"]]
# label: 0=Real, 1=Fake ✅ already correct

print(f"\nFakeNewsNet: {len(fakenews_df)} examples")
print(fakenews_df["label"].value_counts())
```

```python
target_size = len(welfake_df)  # ~72k — use WELFake as anchor

liar_upsampled = resample(
    liar_df,
    replace=True,
    n_samples=target_size,
    random_state=42
)

fakenews_upsampled = resample(
    fakenews_df,
    replace=True,
    n_samples=target_size,
    random_state=42
)
```

```python
combined_df = pd.concat(
    [welfake_df, liar_upsampled, fakenews_upsampled],
    ignore_index=True
)
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\n{'='*50}")
print(f"FINAL COMBINED DATASET")
print(f"Total examples : {len(combined_df)}")
print(f"Label distribution:\n{combined_df['label'].value_counts()}")
print(f"Real: {(combined_df['label']==0).sum()} | Fake: {(combined_df['label']==1).sum()}")
print(f"{'='*50}")
```

```python
print("\nSample from combined dataset:")
for i in [0, 1, 2]:
    row = combined_df.iloc[i]
    print(f"\nLabel: {'Real' if row['label']==0 else 'Fake'}")
    print(f"Text: {row['text'][:120]}...")
```
