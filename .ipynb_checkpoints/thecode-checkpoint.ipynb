```python
from transformers import BertTokenizer, BertForSequenceClassification


tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)
```

```python
def tokenize(batch):
    return tokenizer(
        batch['content'],
        padding='max_length',
        truncation=True,
        max_length=512
    )
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)
```

```python
tokenized_dataset = dataset.map(tokenize, batched=True)
```

```python
print(tokenized_dataset[0])
```

```python
tokenized_dataset = tokenized_dataset.rename_column("label", "labels") #Adjustment
```

```python
tokenized_dataset.set_format(
    type='torch',
    columns=['input_ids', 'attention_mask', 'labels']
)
```

```python
dataset = tokenized_dataset.train_test_split(test_size=0.2)
train_dataset = dataset['train']
test_dataset = dataset['test']
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='the model',
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_strategy="steps",
    logging_steps=200,
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,   # effective train batch = 8
    weight_decay=0.01,
    num_train_epochs=3,
    fp16=True,                       # use only if your GPU/notebook supports it
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True,
    report_to="none",
)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)

    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average="binary", zero_division=0
    )
    acc = accuracy_score(labels, preds)

    return {
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
trainer.save_model("fake_news_bert_model")
tokenizer.save_pretrained("fake_news_bert_model")
```

```python
import torch

def predict_news(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=512
    )

    # ✅ move inputs to GPU
    inputs = {key: val.to(device) for key, val in inputs.items()}

    # ✅ make sure model is on GPU too
    model.to(device)
    model.eval()

    with torch.no_grad():
        outputs = model(**inputs)

    probs = torch.softmax(outputs.logits, dim=1)

    pred = torch.argmax(probs).item()

    label = "Fake" if pred == 1 else "Real"
    confidence = probs[0][pred].item()

    return label, confidence
```

```python
text1 = """
BREAKING: Scientists CONFIRM that drinking coffee cures all diseases instantly!!!
Experts say this miracle discovery is being hidden by big pharmaceutical companies.
Thousands of people have already been cured overnight. Share this before it gets deleted!
"""  #Fake

text2 = """
The Ministry of Health announced on Tuesday that new regulations will be introduced
to improve hospital safety standards. Officials stated that the policy will be
implemented gradually over the next six months, with regular evaluations.
""" #Real

text3 = """
A recent report claims that a secret group of researchers has developed a device
that can control weather patterns globally. While no official confirmation has been
provided, sources suggest governments are already using the technology.
"""#Triky fake

text4 = """
According to Reuters, global oil prices rose by 2% on Monday following supply concerns.
Analysts attribute the increase to ongoing geopolitical tensions affecting production.
"""#Real
```

```python
predict_news(text3)
```

```python
# =========================
# Test your model on examples
# =========================

test_samples = [
    # 🟢 REAL
    "The Federal Reserve announced it will keep interest rates unchanged while monitoring inflation trends and economic growth.",
    "The United Nations held an emergency meeting to discuss climate change policies and international cooperation efforts.",
    "Apple unveiled its latest iPhone model featuring improved battery life and enhanced camera capabilities during its annual event.",

    # 🔴 FAKE
    "Scientists have discovered a way to make humans live up to 300 years by drinking a special herbal tea every day.",
    "A man claims he built a working time machine in his garage and traveled to the year 3025.",
    "Doctors confirm that eating chocolate three times a day completely eliminates the risk of all diseases.",

    # ⚠️ TRICKY
    "Experts suggest that certain natural remedies may help boost the immune system, though more research is needed.",
    "A viral post claims that a new study proves vaccines are harmful, but no scientific evidence supports this claim.",
    "Some people believe that drinking hot water frequently can prevent serious illnesses, although this is not medically proven."
]

# Run predictions
for i, text in enumerate(test_samples, 1):
    label, confidence = predict_news(text)
    
    print(f"\nExample {i}")
    print(f"Text: {text}")
    print(f"Prediction: {label}")
    print(f"Confidence: {confidence:.4f}")
    print("-" * 80)
```
